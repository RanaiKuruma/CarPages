```python
#Imports 
import requests 
from bs4 import BeautifulSoup 
import pandas as pd 
```

```python
#Importing url 
website = "https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D=&list_price_max=&maximum_distance=all&zip="
```

```python
#HTTPS Requests 
response = requests.get(website)
response 
#Soup Object 
soup = BeautifulSoup(response.text, 'lxml')
# soup 
```

```python
#To confirm the total no of postings
results = soup.find_all('div', class_ = 'vehicle-card')
# len(results)
```

```python
#Getting all the data in a single page 
results = soup.find_all('div', class_ = 'vehicle-card')[0]

# results
# len(results)
name = results.find('h2', class_ = 'title').text
# name 
mileage = results.find('div', class_ = 'mileage').text
# mileage 
rating = results.find('span', class_ = 'sds-rating__count').text
review_count = results.find('span' ,class_ = 'sds-rating__link sds-button-link').text
# rating 
# review_count
price = results.find('span', class_ = 'primary-price').text
# price 
dealer_name = results.find('div', {'class': 'dealer-name'}).text.strip()
 
```

```python
#Put everything in a for loop 
results = soup.find_all('div', class_ = 'vehicle-card')
names = []
mileages = []
dealer_names = []
ratings = []
review_counts = []
prices = []

for result in results:
    #Name
    try:
        names.append(result.find('h2', class_ = 'title').text)
    except:
        names.append('n/a')   

    #Mileage 
    try:
        mileages.append(result.find('div', class_ = 'mileage').text)             
    except:
        mileages.append('n/a')  

    #Dealer Name  
    try:
        dealer_names.append(result.find('div', {'class': 'dealer-name'}).text.strip())             
    except:
        dealer_names.append('n/a')     

    #Rating
    try:
        ratings.append(result.find('span', class_ = 'sds-rating__count').text)             
    except:
        ratings.append('n/a')                   

    #Review Count 
    try:
        review_counts.append(result.find('span' ,class_ = 'sds-rating__link sds-button-link').text)             
    except:
        review_counts.append('n/a')                   

    #Price 
    try:
        prices.append(result.find('span', class_ = 'primary-price').text)
    except:
        prices.append('n/a')          
# names   
# mileages  
# dealer_names
# ratings
# review_counts
# prices 
```

```python
#Put everything in a data frame 
car_dealer = pd.DataFrame({'Name':names, 'Mileage':mileages, 'Dealer-Name':dealer_names, 'Ratings': ratings, 'Rating Counts': review_counts, 'Price': prices})
# car_dealer 
```

```python
#Data Cleaning (Checking out on a single page)
car_dealer['Rating Counts'] = car_dealer['Rating Counts'].apply(lambda x:x.strip('reviews'))
# car_dealer
```

```python
#Output in excel (Sinlge Page)
car_dealer.to_excel('single_car_page.xlsx', index = False)
```

```python
#Pagination
names = []
mileages = []
dealer_names = []
ratings = []
review_counts = []
prices = []

for i in range(1,11):
    #Website in a variable 
    website = 'https://www.cars.com/shopping/results/?page=' + str(i) + '&page_size=20&list_price_max=&makes[]=mercedes_benz&maximum_distance=all&models[]=&stock_type=cpo&zip='

    #Request to a website 
    response = requests.get(website)

    #Soup 
    soup = BeautifulSoup(response.text, 'lxml')
    
    #Results 
    results = soup.find_all('div', class_ = 'vehicle-card')


    for result in results:
        #Name
        try:
            names.append(result.find('h2', class_ = 'title').text)
        except:
            names.append('n/a')   

        #Mileage 
        try:
            mileages.append(result.find('div', class_ = 'mileage').text)             
        except:
            mileages.append('n/a')  

        #Dealer Name  
        try:
            dealer_names.append(result.find('div', {'class': 'dealer-name'}).text.strip())             
        except:
            dealer_names.append('n/a')     

        #Rating
        try:
            ratings.append(result.find('span', class_ = 'sds-rating__count').text)             
        except:
            ratings.append('n/a')                   

        #Review Count 
        try:
            review_counts.append(result.find('span' ,class_ = 'sds-rating__link sds-button-link').text)             
        except:
            review_counts.append('n/a')                   

        #Price 
        try:
            prices.append(result.find('span', class_ = 'primary-price').text)
        except:
            prices.append('n/a')          

```

```python
car_dealer = pd.DataFrame({'Name':names, 'Mileage':mileages, 'Dealer-Name':dealer_names, 'Ratings': ratings, 'Rating Counts': review_counts, 'Price': prices})
#Data Cleaning
car_dealer['Rating Counts'] = car_dealer['Rating Counts'].apply(lambda x:x.strip('()'))
car_dealer
```

```python
#Output in excel 
car_dealer.to_excel('Multiple_car_Pages.xlsx', index = False)
```
